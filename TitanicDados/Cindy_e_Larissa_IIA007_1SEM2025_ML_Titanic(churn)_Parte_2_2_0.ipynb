## **Definindo as amostras de Treino e Teste**

```python
# Definindo a amostra de treino
# X: variáveis independentes (todos menos a coluna alvo)
X = retention_train[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_female']]

# y: variável alvo (o que queremos prever)
Y = retention_train['Survived']
```

## **Criando e treinando o classificador**

```python
# Create the logistic regression classifier
classificador = linear_model.LogisticRegression(solver='liblinear', C=1)

# Train the classifier
classificador.fit(X, Y)
```

```python
# Z: variáveis independentes do dataset de teste
Z = retention_test[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_female']]

classificador.predict(Z)
```

```python
true_labels = retention_test['Survived']
true_labels
```

```python
pred_labels = classificador.predict(Z)

pred_labels
```

## **Avaliando a qualidade do modelo**

### *Criando a matriz de confusão*

```python
# Create confusion matrix

confusion_mat = confusion_matrix(true_labels, pred_labels)
```

### *Gerando o classification report*

```python
# Classification report
targets = ['Morreu', 'Sobreviveu']
print("Relatório de Classificação:\n")
print('\n', classification_report(true_labels, pred_labels, target_names=targets))
```

## Plotando a matrix de confusão

```python
import seaborn as sns
```

```python
# Plotando a Matriz de Confusão
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues', xticklabels=['Morreu', 'Sobreviveu'], yticklabels=['Morreu', 'Sobreviveu'])
plt.xlabel('Previsto')
plt.ylabel('Real')
plt.title('Matriz de Confusão')
plt.show()
```

# **CONCLUSÃO**

O modelo utilizado para prever a sobrevivência dos passageiros do Titanic apresentou uma acurácia de 80%, indicando que 80% das previsões foram corretas. Esse é um bom desempenho para esse tipo de problema, considerando os desafios comuns de desbalanceamento e variabilidade nos dados históricos.

Classe "Morreu":

Precisão (0.82): O modelo foi capaz de identificar corretamente, com 82% de acerto, os passageiros que realmente morreram.

Revocação (0.87): Entre todos que realmente morreram, 87% foram corretamente identificados como tal.

F1-score (0.84): Um equilíbrio sólido entre precisão e revocação.

Classe "Sobreviveu":

Precisão (0.77): Dos que foram classificados como sobreviventes, 77% realmente sobreviveram.

Revocação (0.69): O modelo identificou corretamente 69% dos que de fato sobreviveram.

F1-score (0.73): Um desempenho um pouco inferior à classe "Morreu", indicando maior dificuldade do modelo em identificar corretamente os sobreviventes.

Macro e Weighted Averages:
Ambos os valores de macro avg e weighted avg para precisão, revocação e F1-score estão em torno de 0.78 a 0.80, o que confirma um desempenho equilibrado, embora com leve vantagem na classificação da classe majoritária ("Morreu").

Em resumo,o modelo tem um bom desempenho geral, com uma acurácia de 80%. No entanto, ele é ligeiramente melhor em identificar passageiros que morreram do que aqueles que sobreviveram. Isso pode estar relacionado a um desbalanceamento dos dados, já que há mais exemplos da classe "Morreu" (549) do que da classe "Sobreviveu" (342). Melhorias podem ser alcançadas com técnicas de balanceamento de classes, ajuste de hiperparâmetros ou uso de algoritmos mais robustos.
